```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy.signal import convolve2d
```

```python
img_path = "profile_pic_2020_1000x1000.png"
kernel_width = 50

def make_box_kernel(L, v):
    kernel = np.ones((kernel_width,kernel_width)) * -v
    min = int((kernel_width - L) / 2)
    max = int((kernel_width + L) / 2)
    kernel[min:max, min:max] = 1
    return kernel
```

```python
img = mpimg.imread(img_path)
plt.imshow(img)
plt.show()
```

```python
if img.ndim == 3:
    img = img.mean(axis=2)

kernels = [
    make_box_kernel(40, 1),
    make_box_kernel(25, 1),
    make_box_kernel(25, 5),
    make_box_kernel(25, 0.5),
]

for kernel in kernels:
    plt.imshow(kernel, cmap="gray")
    plt.show()

    out = convolve2d(img, kernel, mode='same')
    plt.imshow(out, cmap="gray")
    plt.show()
```
